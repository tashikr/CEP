```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# ===== Step 1: Load & Preprocess =====
df = pd.read_csv("form_csv.csv")

# Drop irrelevant columns
df_clean = df.drop(columns=["Timestamp", "Enter your name", "Email Address"])

# Imputer for missing values
imputer = SimpleImputer(strategy="most_frequent")
df_imputed = pd.DataFrame(imputer.fit_transform(df_clean), columns=df_clean.columns)

# Label encoding
le_dict = {}
for col in df_imputed.columns:
    if df_imputed[col].dtype == "object":
        le = LabelEncoder()
        df_imputed[col] = le.fit_transform(df_imputed[col])
        le_dict[col] = le

# ===== Step 2.1: Age Distribution (Remove unrealistic ages > 100) =====
plt.figure(figsize=(8,5))

# Convert to numeric and filter ages between 0 and 100
ages = pd.to_numeric(df_imputed["What is your age?"], errors="coerce")
ages = ages[(ages >= 0) & (ages <= 100)]

ages.hist(bins=15, edgecolor="black")
plt.title("Age Distribution of Respondents")
plt.xlabel("Age")
plt.ylabel("Number of Participants")
plt.grid(axis="y", alpha=0.3)
plt.show()

# ===== Step 2.2: Gender Distribution (Pie Chart) =====
plt.figure(figsize=(6,6))
gender_counts = df["Gender"].value_counts()
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    colors=["#66b3ff","#ff9999","#99ff99"]
)
plt.title("Gender Distribution of Respondents")
plt.show()
```

```python

# ===== Step 2.3: Screen Time Distribution =====
plt.figure(figsize=(8,5))
screentime_counts = df["How much screentime do you average per day?"].value_counts()
screentime_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Daily Screen Time Distribution")
plt.xlabel("Screen Time")
plt.ylabel("Number of Participants")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.3)
plt.show()

# ===== Step 2.4: Correlation Heatmap (without seaborn) =====
plt.figure(figsize=(10,7))
correlation_matrix = df_imputed.corr()

plt.imshow(correlation_matrix, cmap="coolwarm", interpolation="nearest")
plt.colorbar(label="Correlation")

# Add labels
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)

plt.title("Correlation Heatmap: Social Media & Mental Health Survey")
plt.show()

# ===== Step 2.5: Relationship Example: Screen Time vs Anxiety/Stress =====
plt.figure(figsize=(8,5))
cross_tab = pd.crosstab(
    df["How much screentime do you average per day?"],
    df["After spending time on social media, do you feel"]
)

cross_tab.plot(kind="bar", stacked=True, figsize=(10,6), colormap="viridis", edgecolor="black")
plt.title("Screen Time vs Reported Feelings After Social Media Use")
plt.xlabel("Daily Screen Time")
plt.ylabel("Number of Participants")
plt.xticks(rotation=45)
plt.show()
```

```python
# ===== Step 2.6: Scatter Plot - Screen Time vs Comparison =====

plt.figure(figsize=(8,5))

# Use the encoded version (df_imputed) so values are numeric
x = df_imputed["How much screentime do you average per day?"]
y = df_imputed["How often do you compare yourself to others on social media"]

plt.scatter(x, y, alpha=0.6, c="teal", edgecolors="black")

plt.title("Screen Time vs Comparison to Others")
plt.xlabel("Screen Time (encoded)")
plt.ylabel("Comparison Frequency (encoded)")
plt.grid(alpha=0.3)
plt.show()
```
